```python
#名前(name)と年齢(age)の属性を持つHumanクラスを作成してください。
#nameとageを標準出力(print)するメソッド(printinfo)を追加してください。
#Humanクラスのインスタンスは、変数に代入してプログラム内で使用してください。

class Human:
    # コンストラクタを定義する
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # メソッドを定義する
    def printinfo(self):
          print(f"名前: {self.name}, 年齢: {self.age}")

# インスタンス化する
human = Human("田原賢", 29)

human.printinfo()
```
