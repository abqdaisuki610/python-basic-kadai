```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数numに0～30までのランダムな整数を代入する
var = random.randint(0,30)

# 3の倍数（3で割ったあまりが0）かつ5の倍数（5で割ったあまりが0）でFizzBuzz    
if var % 3 == 0 and var % 5 == 0:
    print("FizzBuzz")
# 3の倍数（3で割ったあまりが0）でFizz
elif var % 3 == 0:
    print("Fizz")
# 5の倍数（5で割ったあまりが0）でBuzz
elif var % 5 == 0:
    print("Buzz")

else:
    print(var)
```
